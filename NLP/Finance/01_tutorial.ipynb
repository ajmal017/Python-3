````python
def get(tickers, startdate, enddate):
    def data(ticker):
        return (pdr.get_data_google(ticker, start=startdate, end=enddate))
    datas = map (data, tickers)
    return(pd.concat(datas, keys=tickers, names=['Ticker', 'Date']))

tickers = ['AAPL', 'MSFT', 'IBM', 'GOOG']
all_data = get(tickers, datetime.datetime(2006, 10, 1), datetime.datetime(2012, 1, 1))
````

````python
all_data.head()
````

````python
daily_close_px = all_data[['Close']].reset_index().pivot('Date', 'Ticker', 'Close')
````

````python
daily_close_px.head()
````

````python
daily_pct_change = daily_close_px.pct_change()
````

````python
daily_pct_change.hist(bins=20, sharex=True, figsize=(12,8))
````

kde = Kernel Density Estimate plot

````python
pd.scatter_matrix(daily_pct_change, diagonal='kde', alpha=0.5,figsize=(12,12))
````
